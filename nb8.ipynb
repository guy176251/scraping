```python
import re

pattern = re.compile(' \(can only live in Mushroom\)|(Loves|Likes|Dislikes) |\\n|None')
clean_text = lambda elem: pattern.sub('', elem.text_content())
multi_npc = 'small span span span a'

raw_lists = [[[clean_text(element) for element in all_elems(td, multi_npc)]
              if all_elems(td, multi_npc)
              else [clean_text(td)]
              
              for td in all_elems(tr, 'td')]
             for tr in all_elems(npc_pref_elem, 'tbody tr')][2:]

def empty_checker(field, biome=False):
    for thing in field:
        if not thing:
            return None
        
    return field[0] if biome else field

love = 10
like = 5
dislike = -5
hate = -10

npc_pref_dict = {
    npc[0]: {
        'biome': {
            like: empty_checker(biome_like, True),
            dislike: empty_checker(biome_dislike, True)
        },
        'npc': {
            love: empty_checker(npc_love),
            like: empty_checker(npc_like),
            dislike: empty_checker(npc_dislike),
            hate: empty_checker(npc_hate)
        }
    }
    
    for npc, biome_like, biome_dislike, npc_love, npc_like, npc_dislike, npc_hate
    in raw_lists
}

npc_pref_dict
```

```python
npc_pref_values = {}

for npc1 in npc_pref_dict:
    
    npc_pref_values[npc1] = {}
    
    for npc2 in npc_pref_dict:
        total = []
        
        if npc1 == npc2:
            continue
        
        for field in (npcs := npc_pref_dict[npc1]['npc']):
            if npcs[field] and npc2 in npcs[field]:
                total.append(field)
        
#         biomes_1 = npc_pref_dict[npc1]['biome']
#         biomes_2 = npc_pref_dict[npc2]['biome']
        
#         if biomes_1[dislike] == biomes_2[like] or biomes_2[dislike] == biomes_1[like]:
#             total.append(dislike)
#         elif biomes_1[like] == biomes_2[like]:
#             total.append(like)
        
        if (summed := sum(total)) > 0:
            npc_pref_values[npc1][npc2] = summed
            
# for npc in sorted(npc_pref_values, key=lambda x: len(npc_pref_values[x])):
#     print(f'{npc}\n   ',
#           '\n    '.join(f'{n}: {npc_pref_values[npc][n]}' for n in npc_pref_values[npc]),
#           '\n')

shown = []
for npc_out in (outer := npc_pref_values):

    if npc_out in shown:
        continue

    for npc_in in (inner := outer[npc_out]):

        if npc_in == npc_out:
            continue

        if npc_in in shown:
            continue

        if npc_out in outer[npc_in]:
            print(f'{npc_in} ({outer[npc_in][npc_out]}) and {npc_out} ({inner[npc_in]})')
            shown += [npc_in, npc_out]
```
